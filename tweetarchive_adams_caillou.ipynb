```python
search_words = "Stacey Abrams" +  " -filter:retweets"
date_since = "2021-01-06"
```

```python
tweets = tw.Cursor(api.search, 
                           q=search_words,
                           lang="en",
                           since=date_since).items(2000)
```

```python
stacey_tweets = [[tweet.favorite_count, tweet.retweet_count, tweet.text, tweet.user.location] for tweet in tweets]
```

```python
adams_df = pd.DataFrame(data=stacey_tweets, 
                    columns=["Likes", "Retweets", "Text", "User Location"])
```
